### Question 2

**2.** Which columns had no missing values? Provide a set of column names that have no missing values.

```python
no_nulls = set()
for col in df.columns:
    if df[col].isnull().sum() == 0:
        no_nulls.add(col)

display.HTML(t.no_null_cols(no_nulls))
```

### Question 3

**3.** Which columns have the most missing values?  Provide a set of column names that have more than 75% if their values missing.

```python
most_missing_cols = set(df.columns[df.isnull().mean() > 0.75])#Provide a set of columns with more than 75% of the values missing

t.most_missing_cols(most_missing_cols)
```

Let's see which county's developer earns more

```python
pd.DataFrame(df.query(" Professional == 'Professional developer' ")).groupby(['Country', 'Professional'])['Salary'].mean().sort_values(ascending=False)[:10]
```

That's interesting Bermuda's developer earns more than any other country.Let's find out some more interesting things about the dataset.

Let's see how many professional developer's has computer science major. This will tell us about how many non-cs people come into developer line.

```python
cs_devs = df[(df.Professional == "Professional developer") & (df.MajorUndergrad == "Computer science or software engineering")]['Professional'].count()
print("%age of people working as professional developer have cs major {}".format(round((cs_devs/df.shape[0])*100, 2)))
```

Which country has maximum number of data scientists present.

```python
pd.DataFrame(df[df.NonDeveloperType == "Data scientist"][["Country", "NonDeveloperType"]]).groupby('Country').apply(
    lambda x: x.NonDeveloperType.count()).sort_values(ascending=False)[:10].plot(kind='bar')
```
